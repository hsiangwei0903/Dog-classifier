```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os

import torchvision.transforms as transforms
import matplotlib.pyplot as plt
  
# define custom transform function
transform = transforms.Compose([
    transforms.ToTensor()
])

count = 0
mean, std = np.array([0, 0, 0]), np.array([0, 0, 0])
for breed in os.listdir('/content/drive/MyDrive/Capstone/wyze/test'):
  for img_path in os.listdir(f'/content/drive/MyDrive/Capstone/wyze/test/{breed}'):
    img = Image.open(f'/content/drive/MyDrive/Capstone/wyze/test/{breed}/{img_path}')
    img_tr = transform(img)
    # print(np.asarray(img_tr.mean([1,2])))
    mean = np.add(np.asarray(img_tr.mean([1,2]))[:3], mean)
    std = np.add(np.asarray(img_tr.std([1,2]))[:3], std)
    count += 1

print(mean / count)
print(std / count)
```
